## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
import scipy.stats as stats

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data with the panda library 
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data_df=pd.merge(mouse_metadata,study_results,on="Mouse ID",how="left")
combined_data_df.head()
```

```python
# Checking the number of mice.
mice_number=len(combined_data_df["Mouse ID"].unique())
mice_number
           
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dupe_number=combined_data_df[combined_data_df.duplicated(subset = ["Mouse ID", "Timepoint"])]
dupe_number
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df =combined_data_df.loc[combined_data_df["Mouse ID"]!="g989"]
clean_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
new_mice_number=len(clean_df["Mouse ID"].unique())
new_mice_number
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

drug_regimen=clean_df.groupby(["Drug Regimen"])

mean_value=drug_regimen["Tumor Volume (mm3)"].mean()

median_value=drug_regimen["Tumor Volume (mm3)"].median()

variance=drug_regimen["Tumor Volume (mm3)"].var()

standard_deviation=drug_regimen["Tumor Volume (mm3)"].std()

sem_value=drug_regimen["Tumor Volume (mm3)"].sem()

# Create new DataFrame
single_summary_df = pd.DataFrame({"Median Value": median_value,
                                           "Average Volume":mean_value,
                                           "Variance": variance,
                                           "Standard Deviation": standard_deviation,
                                            "Sem": sem_value})
single_summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
drug_regimen.agg({'Tumor Volume (mm3)':["median", "mean", "var", "std", "sem"]})
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:Capomulin, Ramicane, Infubinol, & Ceftamin
cap_df = clean_df.loc[clean_df["Drug Regimen"] == "Capomulin",:]
ram_df = clean_df.loc[clean_df["Drug Regimen"] == "Ramicane", :]
inf_df = clean_df.loc[clean_df["Drug Regimen"] == "Infubinol", :]
ceft_df = clean_df.loc[clean_df["Drug Regimen"] == "Ceftamin", :]
cap_df.head()

# Start by getting the last (greatest) timepoint for each mouse
capmax = cap_df.groupby('Mouse ID')['Timepoint'].max()
capmaxvol = pd.DataFrame(capmax)

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
capmaxmerge = pd.merge(capmaxvol, clean_df, on=("Mouse ID","Timepoint"),how="left")
capmaxmerge.head(5)
```

```python
A1 = clean_df.pivot_table(index = "Mouse ID", values=["Timepoint"], aggfunc="max")
#A1.head()
A2 = pd.merge(A1, clean_df, how = "left", on=["Mouse ID", "Timepoint"])
A2.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_list=["Capomulin","Ramicane", "Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in drug_list:
    drug_volume=A2.loc[A2["Drug Regimen"]==drug]["Tumor Volume (mm3)"]
    tumor_volume.append(drug_volume)

    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    quartiles = drug_volume.quantile([.25,.5,.75])
    lowerq= quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    
    # Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    
    #print(f"{drug} Outliers: {drug_volume.loc[(drug_volume < lower_bound) | (drug_volume > upper_bound)]}")
    
 # identify outliers
    outliers = drug_volume.loc[(drug_volume < lower_bound) | (drug_volume > upper_bound)]
    # print the found outliers information
    if len(outliers)!= 0:
        print(f"The Drug Regiment {drug} has {len(outliers)} outlier:\n")
        print(outliers[31])
             
    
    
   
    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
# Draw boxplot
fig1, ax1 = plt.subplots()
ax1.boxplot(tumor_volume)

plt.show()
```

## Line and Scatter Plots

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capavg = cap_df.groupby(['Mouse ID']).mean()
plt.scatter(capavg['Weight (g)'],capavg['Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')


plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

corr=round(stats.pearsonr(capavg['Weight (g)'],capavg['Tumor Volume (mm3)'])[0],2)
print(f"The correlation between mouse weight and average tumor volume is {corr}")

model=stats.linregress(capavg['Weight (g)'],capavg['Tumor Volume (mm3)'])
model
```
